```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_performance.csv")
print(df.head())
print(df.describe())
```

Part 1: Starting info

This section of the code imports the required Python libraries, loads the student performance dataset from a CSV file into a pandas DataFrame, and performs an initial inspection of the data. The dataset is displayed using the first few rows to verify that it was read correctly and that each feature contains reasonable values. Summary statistics are then generated to provide an overview of the distribution, range, and central tendencies of all numeric features, including the target variable. This initial exploration helps confirm data integrity and informs later analysis and visualization decisions.

```python
plt.figure()
plt.scatter(df["hours_studied"], df["final_score"])
plt.xlabel("Hours Studied per Week")
plt.ylabel("Final Exam Score")
plt.title("Hours Studied vs Final Exam Score")
plt.show()
```

Part 2 Visualization: Hours Studied vs Final Exam Score

This code creates a scatter plot to visualize the relationship between weekly study hours and the final exam score. Each point represents a single student. The plot helps determine whether increased study time is associated with higher exam scores.

```python
plt.figure()
plt.scatter(df["attendance_rate"], df["final_score"])
plt.xlabel("Attendance Rate (%)")
plt.ylabel("Final Exam Score")
plt.title("Attendance vs Final Exam Score")
plt.show()

```

Part 3 Visualization: Attendance Rate vs Final Exam Score

This section generates a scatter plot comparing student attendance rates to final exam scores. The visualization is used to observe whether students with higher attendance tend to perform better on the final exam.

```python
plt.figure()
plt.scatter(df["previous_score"], df["final_score"])
plt.xlabel("Previous Exam Score")
plt.ylabel("Final Exam Score")
plt.title("Previous Score vs Final Exam Score")
plt.show()
```

Part 4 Visualization: Previous Exam Score vs Final Exam Score

This visualization shows the relationship between a students previous exam score and their final exam score. The purpose is to identify whether prior academic performance is a strong predictor of future performance.

```python
plt.figure()
plt.scatter(df["sleep_hours"], df["final_score"])
plt.xlabel("Sleep Hours per Night")
plt.ylabel("Final Exam Score")
plt.title("Sleep Hours vs Final Exam Score")
plt.show()
```

Part 5 Visualization: Sleep Hours vs Final Exam Score

This scatter plot compares the average number of hours a student sleeps per night with their final exam score. It is used to examine whether sleep duration has a noticeable impact on academic performance.

```python
plt.figure()
plt.scatter(df["practice_tests"], df["final_score"])
plt.xlabel("Practice Tests Taken")
plt.ylabel("Final Exam Score")
plt.title("Practice Tests vs Final Exam Score")
plt.show()
```

Part 6 Visualization: Practice Tests vs Final Exam Score

This plot illustrates the relationship between the number of practice tests completed and the final exam score. The visualization helps evaluate whether additional exam preparation improves final performance.
