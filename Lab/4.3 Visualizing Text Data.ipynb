Create 3 visualizations from text data of your choice.  Each visualization should have at least 1 - 2 sentences explaining both the figure and the interpretation.
You may use any LLM and produce whatever visuals you think best illustrate your data.  

```python
import matplotlib.pyplot as plt
from collections import Counter
import re
```

```python
text_data = """
Networking allows computers to exchange data efficiently.
TCP provides reliable communication while UDP focuses on speed.
Security is essential in modern networked systems.
Servers are designed to handle multiple client connections.
"""
```

```python
words = re.findall(r"[a-zA-Z]+", text_data.lower())
sentences = [s.strip() for s in text_data.split('.') if s.strip()]
```

```python

word_counts = Counter(words)
most_common_words = word_counts.most_common(10)
labels, values = zip(*most_common_words)

plt.figure(figsize=(10, 5))

plt.bar(labels, values)

plt.title("Most Common Words in the Text")
plt.xlabel("Word")
plt.ylabel("Frequency")


plt.xticks(rotation=30, ha="right")
plt.margins(x=0.05)

plt.tight_layout()
plt.show()
```

Visualization 1: Most Common Words Used
This bar chart shows the most frequently occurring words in the selected text.
The visualization highlights key technical terms, indicating the main topics emphasized throughout the document.

```python
word_lengths = [len(word) for word in words]

plt.figure()
plt.hist(word_lengths, bins=8)
plt.title("Distribution of Word Lengths")
plt.xlabel("Word Length")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

Visualization 2: Distribution of the Length of Words
This histogram displays how word lengths are distributed across the text.
Most words fall within a medium length range, suggesting the text balances technical terminology with readability.

```python
sentence_lengths = [len(sentence.split()) for sentence in sentences]

plt.figure()
plt.plot(range(1, len(sentence_lengths) + 1), sentence_lengths, marker='o')
plt.title("Word Count per Sentence")
plt.xlabel("Sentence Number")
plt.ylabel("Number of Words")
plt.tight_layout()
plt.show()
```

Visualization 3: Word Count per Sentence
This line graph compares the number of words used in each sentence.
The variation in sentence length shows a mix of concise statements and more detailed explanations, which improves clarity.
